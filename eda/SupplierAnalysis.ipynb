### Filter to known mismatches to get a sense for how many suppliers are different

```python
difference_mask = (ingredient_pivot["State"]["Boston"] != ingredient_pivot["State"]["Culver City Food Lab"])
non_null_mask = ingredient_pivot["State"].apply(lambda x: x.notna().all(), axis = 1)
ingredient_pivot.loc[difference_mask & non_null_mask]
```
